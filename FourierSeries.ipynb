```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.imshow(plt.imread('equation.jpg'));
plt.axis('off');
plt.title('Equation');
```

```python
def f(x):
    if x >= 0 and x < 1:
        return np.pi * x
    elif x == 1:
        return 0
    else:
        return np.pi * (x-2)
x = np.arange(0,2,0.2)
y = [f(i) for i in x]
```

```python
plt.plot(x, y);
plt.title('Graph of the above equation');
```

```python
plt.plot(x, y, label='Original Function')
n_values = [1,2,5,10]
for i in n_values:
    n = np.arange(1,i).reshape(-1,1)
    y_ = lambda x : 2 * (-(((-1)**n)/n )* np.sin(n * np.pi * x.reshape(1,-1))).sum(axis=0)
    plt.plot(x,y_(x),label=f'Approximation (n={i})')
plt.legend()
```
